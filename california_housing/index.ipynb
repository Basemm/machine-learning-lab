## Correlation
### Correlation Matrix

```python
corr_matrix = X_train.select_dtypes(include=['number']).corr()
corr_matrix
```

### Correlation Against Target Label

```python
full_train_set = X_train.join(y_train)
full_train_set.select_dtypes(include=['number']).corr()["median_house_value"].drop("median_house_value").sort_values(ascending=False)
```

### Correlation Against Synthetic New Features

```python
full_train_set["rooms_per_household"] = full_train_set["total_rooms"] / full_train_set["households"]
full_train_set["population_per_household"] = full_train_set["population"] / full_train_set["households"]

full_train_set.select_dtypes(include=['number']).corr()["median_house_value"].drop("median_house_value").sort_values(ascending=False)
```

### Scatter Plot

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt


numeric_features = full_train_set.select_dtypes(include=['number'])
scatter_matrix(numeric_features, figsize=(24, 18))
plt.show()
```

### Select Features Plot Againt the Target Label

#### Housing Income vs House Value

```python
full_train_set.plot(kind="scatter", x="median_income", y="median_house_value", alpha=0.1)
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.axis([0, 16, 0, 550000])
plt.show()
```

#### Scatter Map Using Coordinates, Population and Median Income

```python
X_train.plot(kind="scatter", x="longitude", y="latitude", alpha=0.8,
         s=X_train["population"]/100, label="population",
         c="median_income", cmap=plt.get_cmap("jet"), colorbar=True,
         figsize=(10,7))
plt.legend()
plt.show()
```
