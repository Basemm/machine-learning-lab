## Sample Data

```python
import matplotlib.pyplot as plt
from common import plot_image


plot_image(mnist.data[60])

plt.show()
```

```python
plt.figure(figsize=(9, 9))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

for idx, img in enumerate(mnist.data[:100]):
  plt.subplot(10, 10, idx + 1)
  plot_image(img)

plt.gcf().set_size_inches(9, 9)
plt.show()
```

## Value Distribution

```python
import pandas as pd


target_df = pd.DataFrame({ "target": mnist.target.tolist() })


target_df.value_counts().sort_index().plot(kind='bar')
```
